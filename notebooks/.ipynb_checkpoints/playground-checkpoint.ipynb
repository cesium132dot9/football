```python
import pandas as pd
import numpy as np 
import scipy as sp
import torch
```

```python
fb_match = pd.read_csv("https://raw.githubusercontent.com/xgabora/Club-Football-Match-Data-2000-2025/refs/heads/main/data/Matches.csv")
fb_match
```

```python
(fb_match["Division"] == "E0").any()
```

```python
prem = fb_match[fb_match["Division"] == "E0"]
prem
```

```python
from datetime import date

prem["MatchDate"].dtype == date
```

```python
prem['MatchDate'] = pd.to_datetime(prem["MatchDate"])

twentyfive_five_three = prem[prem["MatchDate"] == "2025-05-03"]

twentyfour_twentyfive = prem[(prem["MatchDate"] >= "2024-08-16") & (prem["MatchDate"] <= "2025-05-25")]
twentyfour_twentyfive
```

```python
tftf_copy = twentyfour_twentyfive.copy()
```

```python
tftf_copy_W = tftf_copy[tftf_copy["FTResult"].isin(['H', 'A'])]
tftf_copy_W["HigherElo"] = np.where(
    tftf_copy_W["HomeElo"] > tftf_copy_W["AwayElo"], 'H', 
    np.where(tftf_copy_W['HomeElo'] < tftf_copy_W['AwayElo'], 'A', 'E')
)
tftf_copy_W
```

```python
selected_cols = ["MatchDate", 'HomeTeam', 'AwayTeam', 'HomeElo', 'AwayElo', 'HigherElo', 'FTResult']
tftf_copy_W = tftf_copy_W[selected_cols]
tftf_copy_W
```

```python
tftf_copy_W_unexpected = tftf_copy_W[tftf_copy_W['HigherElo'] != tftf_copy_W['FTResult']]
tftf_copy_W_unexpected
```
